# Sentiment Exploration

## Setup

```python
# Setup and libraries.
%matplotlib inline
import datetime as dt
import dateutil.parser
import itertools
import json
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
import seaborn as sns
import sys
```

```python
# Configuration.
vizConf = json.load(open('../../conf/visualizations.json'))
dateConf = json.load(open('../../conf/dates.json'))
orgs = json.load(open('../../conf/organizations.json'))
fbSentimentKeys = json.load(open('../../conf/facebookSentimentKeys.json'))

# Paths.
pathTwConsolidated = '../../../data/twitter/consolidated/'
pathFbConsolidated = '../../../data/facebook/consolidated/'
pathTwSentiment = '../../../data/sentiment/twitter/'
pathFbSentiment = '../../../data/sentiment/facebook/'
```

```python
# Seaborn setup.
# sns.set_style('dark')
sns.set_context({"figure.figsize": (vizConf['display']['width'],vizConf['display']['height'])})

# Retain color information.
snsColorNeg = sns.color_palette()[0]
snsColorPos = sns.color_palette()[2]
```

## Data

### Twitter Data

```python
# Get Twitter sentiment and transform to a DF.
# TODO: get from summarized tweet data.
sentimentCols = ['orgName','id','date','class','aggScoreNeg','aggScorePos','meanScoreNegSig','meanScorePosSig']
reducedSentiments = []
twitterStart = dateutil.parser.parse(dateConf['twitter']['start'])
twitterEnd = dateutil.parser.parse(dateConf['twitter']['end'])
# Traverse all organizations (and their sentiments).
for (org, orgData) in orgs.iteritems():
    name = orgData['name']
    # Load sentiments.
    sentiments = json.load(open(pathTwSentiment+org+'.json'))
    # Load and traverse tweets.
    tweets = json.load(open(pathTwConsolidated + [f for f in os.listdir(pathTwConsolidated) if f.startswith(org)][-1]))
    for t in tweets:
        tId = t['id']
        # Filter.
        tDate = dateutil.parser.parse(t['created_at'])
        if (tDate < twitterStart or tDate >= twitterEnd) or (str(tId) not in sentiments or sentiments[str(tId)] is None): continue
        # Process sentiment and begin a container with sentiment information.
        tweetData = dict((k,v) for (k,v) in sentiments[str(tId)].iteritems() if k in sentimentCols)
        tweetData['id'] = tId
        tweetData['orgName'] = name
        tweetData['date'] = tDate
        reducedSentiments.append(tweetData)
# Compose DF.
tweetSentiments = pd.DataFrame(reducedSentiments)
print 'Tweets classified:',len(tweetSentiments)
# Clean up.
tweetSentiments = tweetSentiments[sentimentCols]
tweetSentiments.reset_index(drop=True, inplace=True)
# Check head.
tweetSentiments.head()
```

```python
orgCounts = tweetSentiments.groupby('orgName').count()[[1]]
orgCounts.columns = ['tweets']
orgCounts
```

### Facebook Data

```python
# Get Facebook sentiment and transform to a DF.
# TODO: get from summarized post data.
sentimentCols = ['orgName','id','date','twitterComparable','contentType','class','aggScoreNeg','aggScorePos','meanScoreNegSig','meanScorePosSig']
reducedSentiments = []
fbStart = dateutil.parser.parse(dateConf['facebook']['start'])
fbEnd = dateutil.parser.parse(dateConf['facebook']['end'])
# Traverse all organizations (and their sentiments).
for (org, orgData) in orgs.iteritems():
    name = orgData['name']
    # Load sentiments.
    sentiments = json.load(open(pathFbSentiment+org+'.json'))
    # Load posts; traverse.
    posts = json.load(open(pathFbConsolidated + [f for f in os.listdir(pathFbConsolidated) if f.startswith(org)][-1]))
    for p in posts:
        pId = p['id']
        # Filter.
        pDate = dateutil.parser.parse(p['created_time'])
        if (pDate < fbStart or pDate >= fbEnd) or (pId not in sentiments or sentiments[pId] is None): continue
        # Evaluate date.
        postInTwitterRange = (pDate >= twitterStart and pDate < twitterEnd)
        # Process sentiment and traverse elements.
        for (contentType, contentSent) in sentiments[pId].iteritems():
            postData = dict((k,v) for (k,v) in contentSent.iteritems() if k in sentimentCols)
            postData['id'] = pId
            postData['date'] = pDate
            postData['orgName'] = name
            postData['contentType'] = contentType
            postData['twitterComparable'] = postInTwitterRange
            reducedSentiments.append(postData)
# Compose DF.
postSentiments = pd.DataFrame(reducedSentiments)
# print 'Posts classified:',len(postSentiments)
# Clean up.
postSentiments = postSentiments[sentimentCols]
postSentiments.reset_index(drop=True, inplace=True)
# Check head.
postSentiments.head()
```

### Consolidated Data

```python
# Generate a comparable dataset for use in general analysis.
allSentiments = postSentiments.ix[
    (postSentiments.twitterComparable == True) & (postSentiments.contentType == 'message'),
    [col for col in postSentiments.columns if col not in ['twitterComparable','contentType']]
].append(tweetSentiments).sort(['orgName','date'])
allSentiments.date = pd.DatetimeIndex(allSentiments.date).normalize()
print 'Rows: %d' % allSentiments.shape[0]
print 'First 10:'
allSentiments.head(n=10)
```

## Exploration

### Twitter Exploration

Compare simple means for all statistics.

```python
twMeans = tweetSentiments.groupby('orgName').agg(np.mean)
twMeans['aggDiff']     = twMeans.aggScorePos - twMeans.aggScoreNeg
twMeans['meanSigDiff'] = twMeans.meanScorePosSig - twMeans.meanScoreNegSig
twMeans
```

Comparisons of `aggScoreNeg` and `aggScorePos` fields by news organization.

```python
# Data and plot.
aggTwMeans = twMeans.sort('aggDiff',ascending=False)
ax = aggTwMeans.iloc[:,[1,0]].plot(kind='bar',color=[snsColorNeg,snsColorPos])
ax.lines[0].remove()
# Title.
ax.set_title('Positivty/Negativity on Twitter by Organization')
ax.set_xlabel('Organization')
# Fix legends.
plt.legend(['Positivity','Negativity'])
pass
```

#### Proportion of Positive/Negative Tweets by Organization

```python
# Extract date-sentiment information.
orgClasses = tweetSentiments[['orgName','class']]
orgClasses['neg'] = (orgClasses['class'] == 'Negative')
orgClasses['pos'] = (orgClasses['class'] == 'Positive')
orgClasses['neu'] = (orgClasses['class'] == 'Neutral')
del orgClasses['class']
# Calculate sums.
orgSents = orgClasses.groupby('orgName').aggregate(np.sum)
orgSents['negProp'] = orgSents.neg / (orgSents.neg+orgSents.pos+orgSents.neu)
orgSents['posProp'] = orgSents.pos / (orgSents.neg+orgSents.pos+orgSents.neu)
# Reorder.
orgSents['posNegDiff'] = orgSents['posProp'] - orgSents['negProp']
orgSents.sort('posNegDiff',ascending=False,inplace=True)
```

```python
# Data and plot.
ax = orgSents.iloc[:,[4,3]].plot(kind='bar', color=[snsColorNeg,snsColorPos])
ax.lines[0].remove()
# Title and labels.
ax.set_title('Proportion of Positive/Negative Tweets by Organization',size='x-large')
ax.set_xlabel('Organization',size='x-large')
# Fix legends.
plt.legend(['Positive Tweets','Negative Tweets'])
pass
```

Nothing really all that interesting above—positivity is more pronounced than above but ordering is still pretty similar.

### Facebook Exploration

Compare single means for all statistics.

```python
fbMeans = postSentiments.groupby(['orgName','contentType']).agg(np.mean)
fbMeans.drop('twitterComparable',axis=1,inplace=True)
fbMeans['aggDiff']     = fbMeans.aggScorePos - fbMeans.aggScoreNeg
fbMeans['meanSigDiff'] = fbMeans.meanScorePosSig - fbMeans.meanScoreNegSig
fbMeans.reset_index(inplace=True,level=1)
fbMeans.head(n=12)
```

```python
# One plot per content type.
for (k,s) in fbSentimentKeys.iteritems():
    # Let's just keep it to agg/name for now. Nevermind!
    # if k not in ('agg','name'): continue
    # Data.
    aggFbMeans = fbMeans.ix[fbMeans.contentType==k,:].sort('aggDiff',ascending=False)
    # Plot.
    ax = aggFbMeans.iloc[:,[2,1]].plot(kind='bar',color=[snsColorNeg,snsColorPos])
    ax.lines[0].remove()
    # Title.
    ax.set_title('Positivty/Negativity on Facebook by Organization: '+s['name']+'\n'+s['description'])
    ax.set_xlabel('Organization')
    # Fix legends.
    plt.legend(['Positivity','Negativity'])
    pass
```

#### Time Trend of Positive/Negative Posts

```python
# Extract date-sentiment information.
dateClasses = allSentiments[['date','class']]
dateClasses['neg'] = (dateClasses['class'] == 'Negative')
dateClasses['pos'] = (dateClasses['class'] == 'Positive')
dateClasses['neu']  = (dateClasses['class'] == 'Neutral')
del dateClasses['class']
# Calculate sums.
dateSents = dateClasses.groupby('date').aggregate(np.sum)
dateSents['negProp'] = -dateSents.neg / (dateSents.neg+dateSents.pos+dateSents.neu)
dateSents['posProp'] = dateSents.pos / (dateSents.neg+dateSents.pos+dateSents.neu)
```

```python
# Data and plot.
ax = dateSents.iloc[:,[4,3]].plot(kind='bar', stacked=True, color=[snsColorNeg,snsColorPos])
ax.lines[0].remove()
# Title and labels.
ax.set_title('Proportion of Positive/Negative Posts by Date')
ax.set_xlabel('Date')
ax.set_xticklabels([l[5:] for l in dateSents.index.to_native_types()])
# Fix legends.
plt.legend(['Positive Posts','Negative Posts'])
# Dimensions.
ax.set_ylim(ax.get_ylim()[0],ax.get_ylim()[1]*1.1)
pass
```

I was hoping for more discernibility across dates, but the proportion of positive/negative-classified posts remained fairly consistent. There does appear to be a slight peak in positivity (and a trough in negativity) centered on the day after the election.
