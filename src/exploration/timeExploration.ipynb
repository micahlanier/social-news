# Time Exploration

This notebook aims to explore posting time trends among organizations and across social networks. For a broad look at all social activity, it uses the prediction dataset.

**All times have been converted to Eastern Standard Time (EST) from UTC).** This may be problematic for international news sources (though those international sources were chosen for their reach in the United States). Stil, it should overall make for the most interpretable data.

## Setup

```python
%matplotlib inline
# Libraries.
import json
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import pickle
import seaborn as sns

# Other settings.
vizConf = json.load(open('../../conf/visualizations.json'))

# Library settings.
pd.options.display.max_columns = 50
```

```python
# Data.
tw = pd.read_csv('../../../data/unmergedPostSummaries/twitter.csv')
fb = pd.read_csv('../../../data/unmergedPostSummaries/facebook.csv')

# Track special columns.
outcomeCols_lin = ['clicks','retweets','favorites','likes','shares','comments']
twSpecificCols = set(tw.columns).difference(fb.columns)
fbSpecificCols = set(fb.columns).difference(tw.columns)
binaryCols = ['weekend_utc','weekend_est','retweeted']
categoricalCols = binaryCols+['org','org_category','medium','agg_sentiment_class','sentiment_class','service','day_of_week_utc','day_of_week_est']

# Calculate oscillations of time for regression.
tw['time_est_sin'] = np.sin(tw['time_est'])
tw['time_est_cos'] = np.cos(tw['time_est'])
fb['time_est_sin'] = np.sin(fb['time_est'])
fb['time_est_cos'] = np.cos(fb['time_est'])
tw['minute_est_sin'] = np.sin(tw['minute_est'])
tw['minute_est_cos'] = np.cos(tw['minute_est'])
fb['minute_est_sin'] = np.sin(fb['minute_est'])
fb['minute_est_cos'] = np.cos(fb['minute_est'])

# Calculate normalized columns. Useful for comparisons.
for o in twSpecificCols.intersection(outcomeCols_lin):
    tw[o+'_normalized'] = tw[o].astype('float') / tw['followers_count'].astype('float')
tw['clicks_normalized'] = tw['clicks'].astype('float') / tw['followers_count'].astype('float')
for o in fbSpecificCols.intersection(outcomeCols_lin):
    fb[o+'_normalized'] = fb[o].astype('float') / fb['account_likes'].astype('float')
fb['clicks_normalized'] = fb['clicks'].astype('float') / fb['account_likes'].astype('float')

# Log some outcomes.
tw['clicks_log'] = np.log(tw['clicks'])
fb['clicks_log'] = np.log(fb['clicks'])

# Organizations.
orgs = json.load(open('../../conf/organizations.json'))
tw['org_name'] = None
fb['org_name'] = None
for org, orgData in orgs.iteritems():
    tw.loc[tw.org == org,'org_name'] = orgData['name']
    fb.loc[fb.org == org,'org_name'] = orgData['name']

# Combine. Not all columns overlap.
twfb = tw.append(fb)

# Generate categorical columns.
for c in categoricalCols:
    dummyVals = pd.get_dummies(twfb[c])
    dummyVals.columns = [c+'_'+str(val) for val in dummyVals.columns]
    twfb.merge(dummyVals.iloc[:,1:], left_index=True, right_index=True)
    if c not in fbSpecificCols:
        dummyVals_tw = pd.get_dummies(tw[c])
        dummyVals_tw.columns = [c+'_'+str(val) for val in dummyVals_tw.columns]
        tw = tw.merge(dummyVals_tw.iloc[:,1:], left_index=True, right_index=True)
    if c not in twSpecificCols:
        dummyVals_fb = pd.get_dummies(fb[c])
        dummyVals_fb.columns = [c+'_'+str(val) for val in dummyVals_fb.columns]
        fb = fb.merge(dummyVals_fb.iloc[:,1:], left_index=True, right_index=True, copy=False)
```

```python
# Load prior regression models.
lm_clicks_tw = pickle.load(open('../../../data/models/clicks/lm_clicks_tw.pkl'))
lm_clicks_fb = pickle.load(open('../../../data/models/clicks/lm_clicks_fb.pkl'))
```

## Integrating with Predictive Model

There are obvious differences between organizations on social media, and those differences might have implications for click performance. Let's look at how different timings appear to affect log(click) statistics.

All else being equal, weekend posts appear to outperform weekday posts. This may be due to a lack of competition from other news sources. And it cannot apply to all news: breaking news can obviously not be delayed until a "theoretically" favorable date, but some material (e.g., features) could be delayed, or it could be reposted on those days. Still, there is evidence that organizations are "leaving clicks on the table" by not posting frequently on weekends. This analysis could be enriched by topic modeling or analysis of breaking news, in order to indicate what *kind* of news performs better on weekends.

### Time of Day

Timing is a more complex issue: the oscillations by time cannot be described discretely. To handle that, generate very small time increments for prediction.

```python
# Generate time increments.
timeIncrements = np.arange(0.0,24.0,0.001)

# Store coefficients.
twTimeCoefficients = lm_clicks_tw.coef_[0:5]
fbTimeCoefficients = lm_clicks_fb.coef_[0:5]

# Calculate log outcomes.
twTimeLogOutcomes = timeIncrements*twTimeCoefficients[0] + np.sin(timeIncrements)*twTimeCoefficients[1] + np.cos(timeIncrements)*twTimeCoefficients[2] + np.power(timeIncrements,2)*twTimeCoefficients[3] + np.power(timeIncrements,3)*twTimeCoefficients[4]
fbTimeLogOutcomes = timeIncrements*fbTimeCoefficients[0] + np.sin(timeIncrements)*fbTimeCoefficients[1] + np.cos(timeIncrements)*fbTimeCoefficients[2] + np.power(timeIncrements,2)*fbTimeCoefficients[3] + np.power(timeIncrements,3)*fbTimeCoefficients[4]

# Plot for Twitter.
plt.plot(timeIncrements,twTimeLogOutcomes)
plt.title('Link Performance by Time of Day, Twitter',size='xx-large')
plt.xlabel('Hour',size='x-large')
plt.ylabel('$\Delta$ Log(Clicks), All Else Equal',size='x-large')
plt.xlim((0,timeIncrements.max()))
plt.xticks(range(0,25,2))
plt.show()

# Plot for Facebook.
plt.plot(timeIncrements,fbTimeLogOutcomes)
plt.title('Link Performance by Time of Day, Facebook',size='xx-large')
plt.xlabel('Hour',size='x-large')
plt.ylabel('$\Delta$ Log(Clicks), All Else Equal',size='x-large')
plt.xlim((0,timeIncrements.max()))
plt.xticks(range(0,25,2))
plt.show()
```

As an example, consider a Facebook post at 6am vs 6pm.

```python
# Get log values.
amOutcome = fbTimeLogOutcomes[timeIncrements == 6.000]
pmOutcome = fbTimeLogOutcomes[timeIncrements == 18.000]
# Calculate and print.
print 'AM Value:',amOutcome
print 'PM Value:',pmOutcome
print 'Proportional difference in clicks:',pmOutcome-amOutcome 
```

Values above must be interpreted relatively: Twitter items posted in the lowest trough (around 11am) show predicted clicks approximatley $\frac{1}{2}$ of those predicted around 11pm. Note that the Twitter model does not look *great*, given that it does not oscillate across days. Facebook looks much better.

This analysis may be complicated by time zone issues: the local Twitter maximum could intuitively be caused by the success of European news organizations in publishing content in their local time zones. The relationship is not a clearly oscillating one, so further feature engineering might be a good idea.

Still, this strategy illustrates the limitations of publishing strictly during "business hours" and ignoring the late-day audience. Some publishers may be better-served by delaying their "ramp up" posts and posting more content after 12pm. Facebook shows a much cleaner relationship, and indicates that peak publishing hours are between 12pm and 9pm.

There's little reason to conclude that the top-of-the-hour posting strategy is problematic. Good to know!

## Duplicate Posts

In order to maximize clicks at certain times, it is important to consider the types of content that organizations post during off-peak times: weekends and nights. Specifically, are publishers posting duplicate content then?

```python
# Container for URL frequency.
urlFreq = dict()

# Traverse every record.
for _, r in twfb.iterrows():
    urls = json.loads(r.urls)
    for url in urls:
        urlFreq[url] = (urlFreq.get(url) or 0) + 1
```

```python
# Append information to tw dataframe.
# Values to track: number of links, distinct postings.
numLinks = []
distinctPostings = []
# Traverse every tweet.
for _, r in tw.iterrows():
    urls = json.loads(r.urls)
    numLinks.append(len(urls))
    tweetDistinctPostings = 0
    for url in urls:
        tweetDistinctPostings += urlFreq.get(url) or 0
    distinctPostings.append(tweetDistinctPostings)
# Append values to DF.
tw['num_links'] = numLinks
tw['distinct_postings'] = distinctPostings
```

```python
# Append information to fb dataframe.
# Values to track: number of links, distinct postings.
numLinks = []
distinctPostings = []
# Traverse every tweet.
for _, p in fb.iterrows():
    urls = json.loads(p.urls)
    numLinks.append(len(urls))
    postDistinctPostings = 0
    for url in urls:
        postDistinctPostings += urlFreq.get(url) or 0
    distinctPostings.append(postDistinctPostings)
# Append values to DF.
fb['num_links'] = numLinks
fb['distinct_postings'] = distinctPostings
```

Bucket observations based on time of day. This is a somewhat arbitrary process: use posts between midnight and 6am. Interestingly, publishers tend to push *more* unique (i.e., never reposted) news in that time. 

```python
# Test 1: time of day.

# Split into two populations: low-value times and higher-value times.
tw['off_peak_time'] = tw.time_est < 6
fb['off_peak_time'] = fb.time_est < 6

# Split into separate dataframes for easier manipulation.
# Restrict to cases of relatively few distinct postings, to eliminate odd cases like homepage posts.
tw_offPeak = tw.ix[(tw.off_peak_time == True)  & (tw.distinct_postings < 20),'distinct_postings']
tw_onPeak  = tw.ix[(tw.off_peak_time == False) & (tw.distinct_postings < 20),'distinct_postings']
tw_diff    = tw_onPeak.mean() - tw_offPeak.mean()
fb_offPeak = fb.ix[(fb.off_peak_time == True)  & (fb.distinct_postings < 20),'distinct_postings']
fb_onPeak  = fb.ix[(fb.off_peak_time == False) & (fb.distinct_postings < 20),'distinct_postings']
fb_diff    = fb_onPeak.mean() - fb_offPeak.mean()
# Also get all values for permutation test.
# tw_truncatedDistinctPostings = tw.ix[tw.distinct_postings < 20,'distinct_postings']
# fb_truncatedDistinctPostings = fb.ix[fb.distinct_postings < 20,'distinct_postings']

# Plot.
barWidth = 0.8
peakLabels = ['Twitter\n12am-6am','Twitter\n6am-12pm','Facebook\n12am-6am','Facebook\n6am-12pm']
peakVals = [tw_offPeak.mean(), tw_onPeak.mean(), fb_offPeak.mean(), fb_onPeak.mean()]
peakColors = [sns.color_palette()[0], sns.color_palette()[0], sns.color_palette()[1], sns.color_palette()[1]]
plt.bar(np.arange(0,4), peakVals, width = barWidth, color=peakColors)

# Title, labels, dimensions.
plt.title('Repost Volume, Time of Day', size='xx-large')
plt.ylabel('Times Posted', size='x-large')
plt.xticks(np.arange(0.0,4.0)+barWidth/2,peakLabels, size='x-large')
plt.xlim((-0.5,4.3))

# Show.
plt.show()
```
