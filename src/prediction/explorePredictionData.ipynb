# Prediction Data Exploration

This notebook contains basic due dilligence on the prediction dataset.

## Setup

```python
# Libraries.
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import pylab
import seaborn as sns
import scipy.stats as stats

# Library settings.
pd.options.display.max_columns = 50
```

```python
# Data.
tw = pd.read_csv('../../../data/unmergedPostSummaries/twitter.csv')
fb = pd.read_csv('../../../data/unmergedPostSummaries/facebook.csv')
```

```python
# Track special columns.
outcomeCols_lin = ['clicks','retweets','favorites','likes','shares','comments']
twSpecificCols = set(tw.columns).difference(fb.columns)
fbSpecificCols = set(fb.columns).difference(tw.columns)
binaryCols = ['weekend_utc','weekend_est','retweeted']
categoricalCols = ['medium','org','org_category','medium','sentiment_class','service','day_of_week_utc','day_of_week_est']
```

```python
# Calculate normalized columns.
for o in twSpecificCols.intersection(outcomeCols_lin):
    tw[o+'_normalized'] = tw[o].astype('float') / tw['followers_count'].astype('float')
tw['clicks_normalized'] = tw['clicks'].astype('float') / tw['followers_count'].astype('float')
for o in fbSpecificCols.intersection(outcomeCols_lin):
    fb[o+'_normalized'] = fb[o].astype('float') / fb['account_likes'].astype('float')
fb['clicks_normalized'] = fb['clicks'].astype('float') / fb['account_likes'].astype('float')

# Keep track of new columns.
outcomeCols_norm = [c+'_normalized' for c in outcomeCols_lin]
```

```python
# Log columns as desired.
tw['clicks_log'] = np.log(tw['clicks'])
fb['clicks_log'] = np.log(fb['clicks'])
tw['clicks_normalized_log'] = np.log(tw['clicks_normalized'])
fb['clicks_normalized_log'] = np.log(fb['clicks_normalized'])

# Keep track of new columns.
outcomeCols_norm_log = ['clicks_normalized_log']
```

```python
# Combine. Not all columns overlap.
twfb = tw.append(fb)
```

```python
# Get master outcome list.
outcomeCols = set(outcomeCols_lin).union(set(outcomeCols_norm)).union(set(outcomeCols_norm_log))
# Print all outcomes.
print 'All outcomes:',
print ', '.join(outcomeCols)
```

## Outcomes

```python
# Verify that we should log the clicks outcome for regression.

# Linear scale.
plt.hist(twfb.ix[twfb.clicks.notnull(),'clicks'], bins=20)
plt.title('Clicks: Linear')
plt.show()

# Log scale.
plt.hist(twfb.ix[twfb.clicks_log.notnull(),'clicks_log'], bins=20)
plt.title('Clicks: Logged')
plt.show()

# Log scale double.
ax = twfb.ix[twfb.clicks_log.notnull(),:].hist(column='clicks_log',by='service')
plt.show()

# QQ plot.
cnl = twfb.ix[twfb.clicks_log.notnull(),'clicks_log']
stats.probplot(cnl, dist="norm", plot=pylab)
pylab.show()
```

```python
# Verify that we should log the normalized clicks outcome for regression.

# Linear scale.
plt.hist(twfb.ix[twfb.clicks_normalized.notnull(),'clicks_normalized'], bins=20)
plt.title('Clicks: Linear (Normalized)')
plt.show()

# Log scale.
plt.hist(twfb.ix[twfb.clicks_normalized.notnull(),'clicks_normalized_log'], bins=20)
plt.title('Clicks: Logged (Normalized)')
plt.show()

# Log scale double.
ax = twfb.ix[twfb.clicks_normalized.notnull(),:].hist(column='clicks_normalized_log',by='service')
plt.show()

# QQ plot.
cnl = twfb.ix[twfb.clicks_normalized_log.notnull(),'clicks_normalized_log']
stats.probplot(cnl, dist="norm", plot=pylab)
pylab.show()
```

Definitely not a perfectly normal fit, but it looks *pretty* good.
