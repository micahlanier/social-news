# Social Flow Analysis

Just a quick and dirty look at how many organizations use social flow, identified by [trib.al](http://www.trib.al/) links in URL history.

```python
# Setup.
# Libraries.
import json
import pandas as pd
from urlparse import urlparse
```

```python
# Get organizations.
orgs = json.load(open('../../conf/organizations.json'))

# Set up trib.al links info container.
tribalData = dict((o,dict({'tribalLinks':0})) for o in orgs.keys())

# Traverse organizations and get URL information.
for org in orgs.keys():
    # Load URLs.
    orgUrls_tw = json.load(open('../../../data/urls/twitter/%s.json' % org))
    orgUrls_fb = json.load(open('../../../data/urls/facebook/%s.json' % org))
    # Combine URLs.
    orgUrls = dict(orgUrls_tw.items() + orgUrls_fb.items())
    # Traverse all URLs and look for trib.al links.
    for urlInfo in orgUrls.values():
        for historyUrl in urlInfo['history']:
            urlParsed = urlparse(historyUrl)
            if urlParsed.netloc == 'trib.al':
                tribalData[org]['tribalLinks'] += 1
                break
    tribalData[org]['allLinks'] = len(orgUrls)
```

```python
tribalData = pd.DataFrame(tribalData).transpose()
tribalData
```
