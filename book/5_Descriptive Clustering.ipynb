#Cluster Analysis#

```python
"""
Calculate proportion of negative, positive, neutral for each observation (news organization)
param[in] inputDF: input dataframe with observations or texts classified as 'Positive', 'Neutral', and 'Negative'
param[in] key: columns will be used as the key for the groupby function  
param[in] countColumn: column that will be used to count the number of positive, neutral, and negative
param[in] columnNamePrefix: name prefix added to column names
"""
def makeSentimentDF(inputDF, key, countColumn, columnNamePrefix):
    result = inputDF.groupby(key).count()[countColumn]
    result = result.unstack()
    result.columns = result.columns.droplevel()
    result['Total'] = result.Negative + result.Positive + result.Neutral
    result['NegativePercent'] = result.Negative/result.Total
    result['PositivePercent'] = result.Positive/result.Total
    result['NeutralPercent'] = result.Neutral/result.Total
    # Rename column names
    result.rename(columns={'NegativePercent':columnNamePrefix+'NegativePercent', 
                           'PositivePercent':columnNamePrefix+'PositivePercent',
                           'NeutralPercent':columnNamePrefix+'NeutralPercent'}, inplace=True)
    return result
```

#Use Elbow Method to determine Optimal K in K Means Clustering Algorithm#

```python
# Adapted from http://stackoverflow.com/questions/6645895/calculating-the-percentage-of-variance-measure-for-k-means
from scipy.cluster.vq import kmeans,vq
from scipy.spatial.distance import cdist
K = range(1,10)
scaled_data = scale(meanMetrics)
reduced_data = PCA(n_components=2).fit_transform(scaled_data)

# scipy.cluster.vq.kmeans
KM = [kmeans(reduced_data,k) for k in K]
centroids = [cent for (cent,var) in KM]   # cluster centroids

# alternative: scipy.spatial.distance.cdist
D_k = [cdist(reduced_data, cent, 'euclidean') for cent in centroids]
cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]
avgWithinSS = [sum(d)/reduced_data.shape[0] for d in dist]
```

```python
kIdx = 3
# elbow curve #adapted from online code
plt.figure(figsize=(10,10)) 
plt.plot(K, avgWithinSS, 'b*-')
plt.plot(K[kIdx], avgWithinSS[kIdx], marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters', size=16, fontweight='bold')
plt.ylabel('Average within-cluster sum of squares', size=16, fontweight='bold')
plt.title('Elbow for Social Metrics KMeans Clustering', size=20, fontweight='bold')
plt.show()
```

From the graph above, a reasonable k for the average social metrics per bitly click data will be k = 4. The Elbow Method's\* premise is that if the marginal improvement in variations explained starts to decrease as we add in more clusters, we would have arrived at a somewhat optimal k. This is very similar to the F-test where if the variations explained by a newly added set of predictors doesn't decrease the sum of squares much, we would not reject the null hypothesis. From this phenomenon, the optimal K can be spotted at the "elbow" of the graph as shown above.

Note\*:[Elbow Method](http://en.wikipedia.org/wiki/Determining_the_number_of_clusters_in_a_data_set) 

```python
# Adapted from http://stackoverflow.com/questions/6645895/calculating-the-percentage-of-variance-measure-for-k-means
from scipy.cluster.vq import kmeans,vq
from scipy.spatial.distance import cdist
K = range(1,10)
scaled_data = scale(sentimentOnly)
reduced_data = PCA(n_components=2).fit_transform(scaled_data)

# scipy.cluster.vq.kmeans
KM = [kmeans(reduced_data,k) for k in K]
centroids = [cent for (cent,var) in KM]   # cluster centroids

# alternative: scipy.spatial.distance.cdist
D_k = [cdist(reduced_data, cent, 'euclidean') for cent in centroids]
cIdx = [np.argmin(D,axis=1) for D in D_k]
dist = [np.min(D,axis=1) for D in D_k]
avgWithinSS = [sum(d)/reduced_data.shape[0] for d in dist]

kIdx = 4
# elbow curve #adapted from online code
plt.figure(figsize=(10,10)) 
plt.plot(K, avgWithinSS, 'b*-')
plt.plot(K[kIdx], avgWithinSS[kIdx], marker='o', markersize=12, 
    markeredgewidth=2, markeredgecolor='r', markerfacecolor='None')
plt.grid(True)
plt.xlabel('Number of clusters', size=16, fontweight='bold')
plt.ylabel('Average within-cluster sum of squares', size=16, fontweight='bold')
plt.title('Elbow for Sentiment KMeans Clustering', size=20, fontweight='bold')
plt.show()
```

From the graph above, a reasonable k for the average Twitter and Facebook text sentiment data will be k = 5.

#K-Means Clustering#

```python
"""
Function used to plot the clusters and their cluster boundaries
param[in] kmeans: kmeans classifier object
param[in] data: input data for k means clustering (does not to be scaled)
param[in] reduced_data: PCA reduced data
param[in] margin: margin for overall plot
param[in] nClusters: number of clusters for kmeans
param[in] customColorMap: custom color map to display clustering boundaries
param[in] annotationAdjustment: list of 2x1 list to adjust plots for annotation
"""
# Adapted from scikit-learn http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_digits.html
def visualizeKMeans(kmeans, data, reduced_data, margin, nClusters, customColorMap, annotationAdjustment):
    # Step size of the mesh. Decrease to increase the quality of the VQ.
    h = .002     # point in the mesh [x_min, m_max]x[y_min, y_max].

    # Plot the decision boundary. For that, we will assign a color to each
    x_min, x_max = reduced_data[:, 0].min() - margin, reduced_data[:, 0].max() + margin
    y_min, y_max = reduced_data[:, 1].min() - margin, reduced_data[:, 1].max() + margin
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh. Use last trained model.
    Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(1)
    plt.figure(figsize=(20,15)) 
    plt.clf()
    plt.imshow(Z, interpolation='nearest',
               extent=(xx.min(), xx.max(), yy.min(), yy.max()),
               cmap=customColorMap,
               aspect='auto', origin='lower')
    
    # Plot points
    Y = kmeans.predict(reduced_data)
    for i in xrange(0,nClusters):
        group = np.where(Y==i)
        groupPC = [reduced_data[c] for c in group]
        plt.scatter(groupPC[0][:,0], groupPC[0][:,1], s = 150, c=colormap[i])
    
    # Annotate graph
    for i, txt in enumerate(data.index):
        plt.annotate(txt, (reduced_data[i, 0] + annotationAdjustment[i][0], 
                           reduced_data[i, 1] + + annotationAdjustment[i][1]), color='b', fontsize=20)
    # Plot the centroids as a white X
    centroids = kmeans.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=169, linewidths=3,
                color='w', zorder=10, edgecolors = 'k')
    plt.title('K-means Clustering (PCA-reduced data)\n'
              'Centroids are marked with black cross', size = 20, fontweight = 'bold')
    plt.xlabel('Principal Component 1', size = 20, fontweight = 'bold')
    plt.ylabel('Principal Component 2', size = 20, fontweight = 'bold')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xticks(())
    plt.yticks(())
    plt.show()
```

```python
"""
Function used to plot the clusters and their cluster boundaries with pre-defined categories
param[in] kmeans: kmeans classifier object
param[in] data: input data for k means clustering (does not to be scaled)
param[in] reduced_data: PCA reduced data
param[in] margin: margin for overall plot
param[in] nClusters: number of clusters for kmeans
param[in] organization_data: pre-defined organization categories
param[in] category: list of unique categories that exist in organization_data
param[in] customColorMap: custom color map to display clustering boundaries
param[in] annotationAdjustment: list of 2x1 list to adjust plots for annotation
"""
# Adapted from scikit-learn http://scikit-learn.org/stable/auto_examples/cluster/plot_kmeans_digits.html
def visualizeKMeansBoundaries(kmeans, data, reduced_data, margin, nClusters, organization_data, category, customColorMap, annotationAdjustment):
    # Step size of the mesh. Decrease to increase the quality of the VQ.
    h = .002     # point in the mesh [x_min, m_max]x[y_min, y_max].

    # Plot the decision boundary. For that, we will assign a color to each
    x_min, x_max = reduced_data[:, 0].min() - margin, reduced_data[:, 0].max() + margin
    y_min, y_max = reduced_data[:, 1].min() - margin, reduced_data[:, 1].max() + margin
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    # Obtain labels for each point in mesh. Use last trained model.
    Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure(1)
    plt.figure(figsize=(20,15)) 
    plt.clf()
    plt.imshow(Z, interpolation='nearest',
               extent=(xx.min(), xx.max(), yy.min(), yy.max()),
               cmap=customColorMap,
               aspect='auto', origin='lower')
    
    
    i = 0
    for groupType in list(set(organization_data[category])):
        group = np.where(organization_data[category]==groupType)
        groupPC = [reduced_data[c] for c in group]
        plt.scatter(groupPC[0][:,0], groupPC[0][:,1], s = 150, c=colormap[i], label=groupType)
        i += 1
    
    # Annotate graph
    for i, txt in enumerate(data.index):
        plt.annotate(txt, (reduced_data[i, 0] + annotationAdjustment[i][0], 
                           reduced_data[i, 1] + + annotationAdjustment[i][1]), color='b', fontsize=20)
    # Plot the centroids as a white X
    centroids = kmeans.cluster_centers_
    plt.scatter(centroids[:, 0], centroids[:, 1],
                marker='x', s=169, linewidths=3,
                color='k', zorder=10)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=24, markerscale=2, labelspacing=0.5)
    plt.title('K-means Clustering (PCA-reduced data)\n'
              'Centroids are marked with black cross', size = 20, fontweight = 'bold')
    plt.xlabel('Principal Component 1', size = 20, fontweight = 'bold')
    plt.ylabel('Principal Component 2', size = 20, fontweight = 'bold')
    plt.xlim(x_min, x_max)
    plt.ylim(y_min, y_max)
    plt.xticks(())
    plt.yticks(())
    plt.show()
```

```python
"""
Function used to run K Means classifier and plot clusters and their cluster boundaries
param[in] data: input data for k means clustering (does not to be scaled)
param[in] nComponents: number of components for PCA
param[in] customColorMap: custom color map to display clustering boundaries
param[in] annotationAdjustment: list of 2x1 list to adjust plots for annotation
param[in] randomState: random State for K Means clustering
"""
def runKMeans(data, nComponents, nClusters, customColorMap, annotationAdjustment, randomState=99):
    #Scale the data
    scaled_data = scale(data)
    reduced_data = PCA(n_components=nComponents).fit_transform(scaled_data)
    
    for n in nClusters:
        kmeans = KMeans(init='k-means++', n_clusters=n, n_init=10, random_state=randomState)
        kmeans.fit(reduced_data)
        visualizeKMeans(kmeans, data, reduced_data, 0.5, n, customColorMap, annotationAdjustment)
```

**Note***: Social metric per bitly click (PBC) is defined as the social metric (likes, retweet, etc)/bitly click.
